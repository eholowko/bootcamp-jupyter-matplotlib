```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
n = 10
cmap=plt.cm.get_cmap('viridis',n)
cmap.colors
y = np.repeat(1,n)
x = np.arange(len(y))
s = 1              
plt.axis('equal')
plt.bar(x, y, s, color=cmap.colors);
```

```python
flights = sns.load_dataset("flights")
sredniPasazerowie = flights["passengers"].groupby(flights["year"]).agg(np.mean)
sredniPasazerowie
```

```python
plt.style.use('seaborn-whitegrid')
ax = plt.subplot()
ax.plot(sredniPasazerowie,'o--b',label = 'średnia roczna liczba pasażerów')
ax.set_xlabel("Rok",color='y')
ax.set_title("Wykres średniej liczby pasażerów od roku",color = 'm')

ax.set_ylabel("Liczba pasażerów",color = 'y')
ax.set_xticks([1954,1957])

ax.legend();
```

```python
tips = sns.load_dataset("tips")
smoker = tips.smoker.value_counts()
smoker
```

```python
plt.style.use('seaborn-pastel')
plt.axis('equal')
plt.title("Palacze")
plt.pie(smoker,startangle=60,labels=smoker.index,explode=(0.3,0),autopct='%1.1f%%',shadow=True)
plt.legend();
plt.savefig("palacze.pdf")
```

```python
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
fig, ax = plt.subplots()
width = 0.75
rects1 = ax.bar(0, smoker[1], width, color='b',yerr = 5)
rects2 = ax.bar(0 + width, smoker[0], width, color='r',yerr=5)
autolabel(rects1)
autolabel(rects2)
```

```python
plt.style.use('seaborn-pastel')
plt.bar(np.arange(len(smoker.values)),smoker.values ,0.8);
```
